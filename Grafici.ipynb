```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define rows and columns
rows = ['Round1', 'Round2', 'Round3', 'Round4', 'Round5']
columns = ['Client1', 'Client2', 'Client3', 'Client4', 'Client5', 'Client6', 'Client7', 'Client8', 'Client9', 'Client10']

# Define losses for each round
losses = {
    'Round2': ['Client2', 'Client6'],
    'Round4': ['Client3'],
    'Round5': ['Client7']
}

# Create the table
table = [['No Loss' for _ in range(len(columns))] for _ in range(len(rows))]

# Fill the table with losses
for round_name, clients in losses.items():
    for client in clients:
        col_index = columns.index(client)
        row_index = rows.index(round_name)
        table[row_index][col_index] = 'Loss'
    

# Transpose the table data
table_data_transposed = np.array(table).T.tolist()


# Display the table using matplotlib
fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
ax.table(cellText=table_data_transposed, rowLabels=columns, colLabels=rows, loc='center')
plt.show()
```

```python
# define plot a graph wit 3 curves
x = [0, 1, 2, 3, 4, 5]
def plot_3_curves(y1, y2, y3, title, ylabel):
    plt.plot(x, y1, label='case 1')
    plt.plot(x, y2, label='case 2')
    plt.plot(x, y3, label='ideal')
    plt.xlabel('time')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# Accuracy variation depending on the number of rounds
title = 'Accuracy value depending on the number of rounds'
ylabel = 'Accuracy'
round1 = [0, 0.75, 0.75, 0.75, 0.75, 0.75]
round2 = [0, 0.77, 0.91, 0.91, 0.91, 0.91]
ideal = [0, 0.79, 0.93, 0.97, 0.98, 0.98]
plot_3_curves(round1, round2, ideal, title, ylabel)

case1 = [0, 0.75, 0.86, 0.86, 0.86, 0.86]
case2 = [0,0.77, 0.90, 0.94, 0.94, 0.94]
plot_3_curves(case1, case2, ideal, title, ylabel)
```
